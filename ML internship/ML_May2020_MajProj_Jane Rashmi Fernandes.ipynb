### 4.1 Age 

```python
sns.distplot(te['Age'],bins=1)
```

Since the distribution is skewed,we replace the missing values by median

```python
te.groupby(['Pclass','Embarked','Sex']).median()
```

```python
def missed_age(cols):
    Age=cols[0]
    Pclass=cols[1]  
    Embarked=cols[2]
    Sex=cols[3]
        
    if pd.isnull(Age):
        if Pclass==1:
            if Embarked=='C':
                if Sex=='female':
                    return 37
                else:
                    return 36.50
            elif Embarked=='Q':
                if Sex=='female':
                    return 33
                else:
                    return 44
            else:
                if Sex=='female':
                    return 33
                else:
                    return 42        
        elif Pclass==2:
            if Embarked=='C':
                if Sex=='female':
                    return 22
                else:
                    return 29.50
            elif Embarked=='Q':
                if Sex=='female':
                    return 30
                else:
                    return 57
            else:
                if Sex=='female':
                    return 29
                else:
                    return 30       
        elif Pclass==3:
            if Embarked=='C':
                if Sex=='female':
                    return 14.25
                else:
                    return 26
            elif Embarked=='Q':
                if Sex=='female':
                    return 20
                else:
                    return 27
            else:
                if Sex=='female':
                    return 23
                else:
                    return 25       
    else:
        return Age
```

```python
te['Age']= te[['Age','Pclass','Embarked','Sex']].apply(missed_age,axis=1)
```

```python
sns.heatmap(te.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

## Step 5: Outliers

 Outliers can be detected using boxplots and scatter plots

```python
plt.figure(figsize=(6,5))
sns.boxplot(y='Age',data=te,palette='winter')
plt.title('Age')
```

```python
te['Age'].describe()
```

```python
te['Age'].quantile(0.75)
```

```python
te['Age'].quantile(0.25)
```

```python
te['Age'].median()
```

```python
IQR=te['Age'].quantile(0.75)-te['Age'].quantile(0.25)
IQR
```

```python
upper_outlierlimit=te['Age'].quantile(0.75)+1.5*IQR
lower_outlierlimit=te['Age'].quantile(0.25)-1.5*IQR
print(upper_outlierlimit)
print(lower_outlierlimit)
```

```python
outliervalues=te[(te['Age']>=upper_outlierlimit)|(te['Age']<=lower_outlierlimit)]
outliervalues.reset_index(inplace=True)
outliervalues.head()
```

```python
ti=outliervalues['index']
ti.head()
```

```python
for i in outliervalues['index']:
    if te['Age'][i]>=upper_outlierlimit:
      te['Age'][i]=te['Age'].quantile(0.95)

    else:
      te['Age'][i]=te['Age'].quantile(0.25)
      
```

Boxplot of Age after outlier treatment

```python
plt.figure(figsize=(6,5))
sns.boxplot(y='Age',data=te,palette='winter')
plt.title('Age')
```
